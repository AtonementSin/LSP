### 4. Modelling

### 4.1 Linear Regression
-----

```python
# Import Library
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
# Menyimpan df yang sudah dibersihkan ke variabel cleaned_df
cleaned_df = pd.read_csv('SAHeart_cleaned.csv')
```

```python
# Import fungsi preprocessing
from sklearn import preprocessing

scaler = preprocessing.RobustScaler()
cleaned_df[['obesity','alcohol']] = scaler.fit_transform(cleaned_df[['obesity','alcohol']])
cleaned_df.head()
```

```python
# Membuat variabel independen (X) dan variabel dependen (y)
X = cleaned_df[['obesity','alcohol']]
y = cleaned_df['chd']

# Pada tahap selanjutnya, data dibagi menjadi 2 bagian:
# Data Latih (Training Data):  70%
# Data Uji (Testing Data): 30%
```

```python
# Membagi dataset ke data latih dan data uji

# Import fungsi train_test_split untuk membagi data dari libray sklearn dan modul model_selection
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=0)
```

```python
# Memanggil fungsi LinearRegression
from sklearn.linear_model import LinearRegression
```

```python
# Memanggil objek LinearRegression ke dalam variabel mlr
lr = LinearRegression()

# Melatih model
lr.fit(X_train, y_train)
```

```python
# Memprediksi model
y_pred_lr = lr.predict(X_test)
# Nilai prediksi
print("Prediction for test set: {}".format(y_pred_lr))
```

```python
# Membandingkan nilai sebenarnya dengan nilai prediksi menggunakan linear regression
lr_diff = pd.DataFrame({'Actual value': y_test, 'Predicted value': y_pred_lr})
lr_diff.head()
```

```python
# Menerapkan threshold pada variabel y_pred_lr ke dalam variabel y_predict_class
y_predict_class = [1 if prob > 0.5 else 0 for prob in y_pred_lr]
```

#### Hasil Akurasi Linear Regression

```python
# Import Library
from sklearn.metrics import accuracy_score

# Menghitung Akurasi
print(f"Accuracy from Linear Regression: {round(accuracy_score(y_test, y_predict_class) * 100, 2)}%")
# Membulatkan nilai akurasi menjadi dua angka di belakang koma
```
